## Trying various other techniques with iris dataset

```python
data = pd.read_csv('iris-dataset.csv')
data.head()
```

#### use a similarity matrix

```python
sns.clustermap(data, cmap='mako', metric='correlation')
```

#### use a different clustering method

```python
sns.clustermap(data, cmap='mako', method='single')
```

#### Ignore the outliers in clustermap

```python
sns.clustermap(data, cmap='mako', robust= True)
```

#### Change the size of figure

```python
sns.clustermap(data, cmap='mako', figsize=(15,12))
```
